```python
import os
import csv
import shutil
import requests
import time
from datetime import datetime
from deepface import DeepFace
```

```python
# Define source images directory and destination directory
images_dir = 'images'
destination_dir = '/Users/dibyendu/Documents/Code/python/sabi-face-scan-1/UTKFace-SABI'
```

```python
# Ensure the destination directory exists
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)
```

```python
def predict_gender_and_race(image_path):
    # Analyze the image
    attributes = DeepFace.analyze(img_path=image_path, actions=['gender', 'race'], enforce_detection=False)

    # Extract gender and race
    gender = 0 if attributes['gender'] == 'Man' else 1  # 0: Male, 1: Female
    race = attributes['dominant_race']

    # Convert race string to integer (mapping: White=0, Black=1, Asian=2, Indian=3, Others=4)
    race_mapping = {"white": 0, "black": 1, "asian": 2, "indian": 3, "other": 4}
    race_code = race_mapping.get(race.lower(), 4)  # Default to 'Others' if not matched

    return gender, race_code
```

```python
# Read the CSV file
csv_file = '/Users/dibyendu/Documents/Code/python/sabi-face-scan-1/Face-Scans-URLs-By-Age.csv'  # replace with your actual CSV file path
with open(csv_file, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        selfie_url = row['selfieUrl']
        age = row['age']
        
        # Get the image file from the URL
        try:
            response = requests.get(selfie_url, stream=True)
            if response.status_code == 200:
                # Extract the image file name from the URL
                image_name = selfie_url.split('/')[-1]
                
                
                
                # Prepare the new file name: 1_0_age_timestamp.jpg.chip.jpg
                timestamp = datetime.now().strftime('%Y%m%d%H%M%S%f')[:-3]  # Current timestamp
                new_image_name = f"1_0_{age}_{timestamp}.jpg.chip.jpg"
                new_image_path = os.path.join(destination_dir, new_image_name)

                # Save the image to the destination directory
                with open(new_image_path, 'wb') as f:
                    f.write(response.content)
                
                # Determine the race and gender of the particular image
                gender, race = predict_gender_and_race(new_image_path)
                print(f"Predicted Gender: {gender}, Predicted Race: {race}")
                
                print(f"Image {image_name} successfully downloaded and renamed to {new_image_name}")
            else:
                print(f"Failed to download image from {selfie_url}. Status code: {response.status_code}")
        except Exception as e:
            print(f"An error occurred: {e}")
```
